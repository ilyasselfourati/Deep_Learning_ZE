## $Keras$ $tuner$ : (Le deuxime modle en optimisant les hyperparamtres)

Dans cette partie, nous allons utiliser **Keras Tuner** pour effectuer une recherche d'hyperparamtres pour un modle de rseau de neurones. Le but de cette parrtie est de trouver les meilleures **valeurs** d'hyperparamtres pour maximiser la prcision (`accuracy`) du modle sur un ensemble de validation.

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, train_size=0.8)
```

```python
from tensorflow import keras
from tensorflow.keras import layers
from kerastuner.tuners import RandomSearch
from sklearn.metrics import accuracy_score

def build_model(hp):
    """
    Fonction construisant et configurant un modle de rseau de neurones 
    en fonction des hyperparamtres spcifis.

    Parameters 
    ----------
    hp : de type HyperParameters fourni par Keras Tuner. Il est utilis pour spcifier les valeurs des hyperparamtres du modle. 

    Returns
    -------
    model : est une instance de Sequential dans Keras, reprsentant le modle de rseau de neurones configur avec les 
    hyperparamtres spcifis.
    """

    model = keras.Sequential()
    model.add(layers.Dense(units=hp.Int('units1', min_value=150, max_value=300, step=20),
                           activation='relu', input_shape=[57], name='hidden_layer1'))
    model.add(layers.Dropout(0.3))
    model.add(layers.Dense(units=hp.Int('units2', min_value=100, max_value=200, step=20),
                           activation='relu', name='hidden_layer2'))
    model.add(layers.Dropout(0.3))
    model.add(layers.BatchNormalization())
    model.add(layers.Dense(1, activation='sigmoid', name='output_layer'))

    model.compile(
        optimizer=keras.optimizers.Adam(
        hp.Choice('learning_rate', values=[1e-2, 1e-1])),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )


    return model

tuner = RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=10,  # Choisissez le nombre d'essais que vous voulez
    executions_per_trial=3,  # Nombre d'excutions par essai pour rduire la variabilit
    directory='my_dir2',
    project_name='keras_tuner2'
)

print('Le rsum des hyperparamtres et de leur espace de recherche :', tuner.search_space_summary() ) # Afficher un rsum de l'espace de recherche des hyperparamtres
#lanons la recherche d'hyperparamtres en utilisant l'ensemble d'entranement (X_train, y_train) et l'ensemble de validation (X_val, y_val).
tuner.search(X_train, y_train,
             epochs=10,
             validation_data=(X_val, y_val)) 

best_model = tuner.get_best_models(num_models=1)[0] #renvoie les meilleurs modles
best_hyperparameters = tuner.get_best_hyperparameters(num_trials=1)[0] #renvoie les meilleurs hyperparamtres
```

<b><font color=blue> Explication : </font></b>

Nous avons utilis keras tuner pour tester plusieurs paramtres et avoir un modle plus performant mais avec le nombre des trials qu'on a mis les modles tester ont donnes des bons rsultats mais le modle de base reste plus performant, ceci est  cause de : 

- `Nombre d'essais insuffisant` : Il est possible que le nombre d'essais (max_trials) que vous avez spcifi pour le tuner ne soit pas suffisant pour explorer de manire exhaustive l'espace des hyperparamtres.

- `Surajustement  l'ensemble de validation` : Le tuner peut choisir des hyperparamtres qui surajustent  l'ensemble de validation. Cela peut entraner de bonnes performances sur l'ensemble de validation, mais des performances infrieures sur de nouveaux exemples.

- `Interaction complexe entre hyperparamtres` : Parfois, l'interaction entre diffrents hyperparamtres peut tre complexe, et le tuner peut ne pas tre en mesure de trouver la meilleure combinaison.
